```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
```

```python
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
train_images, test_images = train_images / 255.0, test_images / 255.0

def train_mlp(hidden_layers, neurons_per_layer, activation_func, epochs=3):
    model = tf.keras.models.Sequential([
        tf.keras.layers.Flatten(input_shape=(28, 28)),
        *[tf.keras.layers.Dense(neurons_per_layer, activation=activation_func) for _ in range(hidden_layers)],
        tf.keras.layers.Dense(10, activation='softmax')
    ])  
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    history = model.fit(train_images, train_labels, epochs=epochs, validation_data=(test_images, test_labels), verbose=0)
    return model, history
```

```python
def plot_training_history(history):
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()

    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()
```

```python
def main():
    model_configurations = [
        {'hidden_layers': 1, 'neurons_per_layer': 64, 'activation_func': 'relu'},
        {'hidden_layers': 1, 'neurons_per_layer': 64, 'activation_func': 'sigmoid'},
        {'hidden_layers': 1, 'neurons_per_layer': 64, 'activation_func': 'tanh'}
    ]

    for config in model_configurations:
        print(f"\nTraining MLP with {config['hidden_layers']} hidden layer(s) and {config['neurons_per_layer']} neurons per layer using {config['activation_func']} activation function:")
        model, history = train_mlp(config['hidden_layers'], config['neurons_per_layer'], config['activation_func'])
        test_loss, test_accuracy = model.evaluate(test_images, test_labels)
        print(f'Test Accuracy: {test_accuracy:.4f}')
        plot_training_history(history)

if __name__ == "__main__":
    main()
```
