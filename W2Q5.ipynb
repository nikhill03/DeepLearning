```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD, Adam, RMSprop
from tensorflow.keras.utils import to_categorical
```

```python
(train_X, train_y), (test_X, test_y) = cifar10.load_data()

train_X = train_X.astype('float32') / 255
test_X = test_X.astype('float32') / 255

train_y = to_categorical(train_y)
test_y = to_categorical(test_y)
```

```python
def create_model():
    model = Sequential([
        Flatten(input_shape=(32, 32, 3)),
        Dense(512, activation='relu'),
        Dense(256, activation='relu'),
        Dense(10, activation='softmax')
    ])
    return model
```

```python
lrs = [0.001, 0.01, 0.1]
opts = [SGD, Adam, RMSprop]
```

```python
for opt in opts:
    for lr in lrs:
        model = create_model()
        optimizer = opt(learning_rate=lr)  # Pass lr as a keyword argument
        model.compile(optimizer=optimizer,  # Use the created optimizer
                      loss='categorical_crossentropy',
                      metrics=['accuracy'])
        hist = model.fit(train_X, train_y, epochs=10, batch_size=128, validation_split=0.2)
        loss, acc = model.evaluate(test_X, test_y)
        train_acc = hist.history['accuracy'][-1]
        val_acc = hist.history['val_accuracy'][-1]
        print(f"{opt.__name__}, LR: {lr}, Train Acc: {train_acc}, Test Acc: {acc}, Generalization Gap: {train_acc - acc}")
```
